El proyecto de esta sesión se codificará y ejecutará en un Jupyter Notebook, así que vamos a explicar brevemente lo que es un notebook. Un entorno de Notebook es una página web que expone un menú simple y celdas en las que puedes ejecutar código Python. Aunque las celdas son entidades separadas que puedes ejecutar individualmente, todas comparten el mismo núcleo Python. Esto significa que todos los nombres que definas en una celda (las variables, funciones, etc.) estarán disponibles en cualquier otra celda.

En términos simples, un núcleo Python es un proceso en el que se ejecuta Python. La página web Notebook es, por lo tanto, una interfaz expuesta al usuario para manejar este núcleo. La página web se comunica con él mediante un sistema de mensajería muy rápido.

Aparte de todas las ventajas gráficas, la belleza de tener un entorno de este tipo radica en la capacidad de ejecutar un script de Python en trozos, y esto puede ser una gran ventaja. ## Preparando los datos
Queremos conseguir la siguiente estructura de datos: vamos a tener una lista de objetos usuario. Cada objeto usuario estará vinculado a una serie de objetos campaña. En Python, todo es un objeto, así que usaremos este término de forma genérica. El objeto usuario puede ser una cadena, un diccionario o cualquier otra cosa.

Una campaña en el mundo de las redes sociales es una campaña promocional que una agencia de medios ejecuta en las redes sociales en nombre de un cliente. Recuerda que vamos a preparar estos datos para que no estén en perfecto estado (pero tampoco estarán tan mal...). En primer lugar, instanciamos el `Faker` que utilizaremos para crear los datos:

```python
fake = faker.Faker()
```

Luego necesitamos nombres de usuario. Queremos 1.000 nombres de usuario únicos, así que hacemos un bucle hasta que el conjunto de nombres de usuario tenga 1.000 elementos. Un conjunto no permite elementos duplicados, por lo que la unicidad está garantizada:

```python
usernames = set()
usernames_no = 1000

# Llene el conjunto con 1000 nombres de usuario únicos
while len(usernames) < usernames_no:
    usernames.add(fake.user_name())
```

A continuación, creamos una lista de usuarios. Cada nombre de usuario se convertirá en un diccionario de usuarios completo, con otros detalles como nombre, sexo y correo electrónico. A continuación, cada diccionario de usuario se vuelca a JSON y se añade a la lista. Esta estructura de datos no es óptima, por supuesto, pero estamos simulando un escenario en el que los usuarios nos llegan así.

```python
def get_random_name_and_gender():
    # 60% de los usuarios serán mujeres
    skew = .6
    male = random.random() > skew
    if male:
        return fake.name_male(), 'M'
    else:
        return fake.name_female(), 'F'

def get_users(usernames):
    users = []
    for username in usernames:
        name, gender = get_random_name_and_gender()
        user = {
            'username': username,
            'name': name,
            'gender': gender,
            'email': fake.email(),
            'age': fake.random_int(min=18, max=90),
            'address': fake.address(),
        }
        users.append(json.dumps(user))
    return users

users = get_users(usernames)
users[:4]
```

La función `get_random_name_and_gender()` es bastante interesante. Utilizamos la función `random.random()` para obtener un número aleatorio distribuido uniformemente entre 0 y 1. Comparamos el número aleatorio con 0.6, para decidir si generamos un nombre masculino o femenino. Esto hace que el 60% de nuestros usuarios sean mujeres.

Observe también la última línea de la celda. Cada celda imprime automáticamente lo que hay en la última línea; por lo tanto, la salida de `users[:4]` es una lista con los cuatro primeros usuarios.

Los analistas utilizan hojas de cálculo todo el tiempo y se les ocurren todo tipo de técnicas de codificación para comprimir la mayor cantidad de información posible en los nombres de las campañas. El formato que hemos elegido es un ejemplo sencillo de esa técnica: hay un código que nos indica el tipo de campaña, luego las fechas de inicio y finalización, luego la edad y el sexo del destinatario y, por último, la moneda. Todos los valores están separados por un guión bajo. El código para generar estos nombres de campaña se puede encontrar en la siguiente celda:

```python
'''
Formato del nombre de la campaña:
InternalType_StartDate_EndDate_TargetAge_TargetGender_Currency
'''

def get_type():
    # Sólo algunos códigos internos sin sentido
    types = ['AKX', 'BYU', 'GRZ', 'KTR']
    return random.choice(types)

def get_start_end_dates():
    duration = random.randint(1, 2 * 365)
    offset = random.randint(-365, 365)
    start = date.today() - timedelta(days=offset)
    end = start + timedelta(days=duration)
    
    def _format_date(date_):
        return date_.strftime("%Y%m%d")
    return _format_date(start), _format_date(end)

def get_age():
    age = random.randrange(20, 46, 5)
    diff = random.randrange(5, 26, 5)
    return '{}-{}'.format(age, age + diff)

def get_gender():
    return random.choice(('M', 'F', 'B'))

def get_currency():
    return random.choice(('GBP', 'EUR', 'USD'))

def get_campaign_name():
    separator = '_'
    type_ = get_type()
    start, end = get_start_end_dates()
    age = get_age()
    gender = get_gender()
    currency = get_currency()
    return separator.join(
        (type_, start, end, age, gender, currency))
```

En la función `get_type()`, utilizamos `random.choice()` para obtener un valor al azar de una colección. `get_start_end_dates()` es un poco más interesante. Calculamos dos enteros aleatorios: la duración de la campaña en días (entre un día y dos años) y un desfase (un número de días entre -365 y 365). Restamos el desplazamiento (como un `timedelta`) de la fecha de hoy para obtener la fecha de inicio y sumamos la duración para obtener la fecha de finalización. Por último, devolvemos representaciones en forma de cadena de ambas fechas.

La función `get_age()` genera un rango de edad objetivo aleatorio, en el que ambos extremos son múltiplos de cinco. Utilizamos la función `random.randrange()`, que devuelve un número aleatorio de un rango definido por los parámetros `start`, `stop` y `step`. Generamos los números aleatorios age (un múltiplo de 5 entre 20 y 46) y diff (un múltiplo de 5 entre 5 y 26). Añadimos diff a edad para obtener el límite superior de nuestro rango de edad y devolvemos una representación de cadena del rango de edad.

El resto de las funciones son sólo algunas aplicaciones de `random.choice()` y la última, `get_campaign_name()`, no es más que un recolector de todas estas piezas del puzzle que devuelve el nombre final de la campaña.

En la siguiente celda escribimos una función que crea un objeto campaña completo:

```python
'''
Datos de campaña:
name, budget, spent, clicks, impressions
'''

def get_campaign_data():
    name = get_campaign_name()
    budget = random.randint(10**3, 10**6)
    spent = random.randint(10**2, budget)
    clicks = int(random.triangular(10**2, 10**5, 0.2 * 10**5))
    impressions = int(random.gauss(0.5 * 10**6, 2))
    return {
        'cmp_name': name,
        'cmp_bgt': budget,
        'cmp_spent': spent,
        'cmp_clicks': clicks,
        'cmp_impr': impressions
    }
```

Usamos algunas funciones diferentes del módulo `random`. `random.randint()` te da un entero entre dos extremos. El problema es que sigue una distribución de probabilidad uniforme, lo que significa que cualquier número del intervalo tiene la misma probabilidad de salir. Para evitar que todos nuestros datos se parezcan, elegimos usar triangular y gauss, para clics e impresiones. Utilizan distribuciones de probabilidad diferentes para que al final tengamos algo más interesante que ver.

Para asegurarnos de que estamos de acuerdo con la terminología: `clicks` representa el número de clics en un anuncio de campaña, `budget` es la cantidad total de dinero asignada a la campaña, `spent` es la cantidad de ese dinero que ya se ha gastado, e `impressions` es el número de veces que la campaña se ha obtenido, como recurso, de su fuente, independientemente del número de `clicks` que se hayan realizado en la campaña. Normalmente, el número de `impressions` es mayor que el número de `clicks`, porque a menudo se ve un anuncio sin hacer clic en él.

Ahora que tenemos los datos, es hora de ponerlos todos juntos:

```python
def get_data(users):
    data = []
    for user in users:
        campaigns = [get_campaign_data()
                    for _ in range(random.randint(2, 8))]
        data.append({'user': user, 'campaigns': campaigns})
    return data
```

Como puede ver, cada elemento de datos es un diccionario con un usuario y una lista de campañas asociadas a ese *user*.

### Descomponer el nombre de la campaña
Ahora es el momento de aumentar la complejidad. En primer lugar, queremos deshacernos de ese horrible nombre de campaña (`cmp_name`). Tenemos que descomponerlo en partes y poner cada parte en su propia columna dedicada. Para ello, utilizaremos el método `apply()` del objeto `Series`.

La clase `pandas.core.series.Series` es una potente envoltura alrededor de un array (piensa en él como una lista con capacidades aumentadas). Podemos extraer un objeto `Series` de `DataFrame` accediendo a él de la misma forma que hacemos con una clave en un diccionario, y podemos llamar a `apply()` sobre ese objeto `Series`, que llamará a una función dada sobre cada elemento de la `Serie` y devolverá una nueva `Serie` con los resultados. Componemos el resultado en un nuevo `DataFrame`, y luego unimos ese `DataFrame` con `df`.

Dentro de `unpack_campaign_name()`, dividimos el nombre de la campaña en partes. Utilizamos `arrow.get()` para obtener un objeto de fecha adecuado a partir de esas cadenas, y luego devolvemos los objetos. Un vistazo rápido a las tres primeras filas revela:

```python
def unpack_campaign_name(name):
    '''
    Método muy optimista, asume que los datos en el nombre de la campaña están siempre en buen estado
    '''
    type_, start, end, age, gender, currency = name.split('_')
    start = arrow.get(start, 'YYYYMMDD').date()
    end = arrow.get(end, 'YYYYMMDD').date()
    return type_, start, end, age, gender, currency

campaign_data = df['cmp_name'].apply(unpack_campaign_name)
campaign_cols = [
    'Type', 'Start', 'End', 'Target Age', 'Target Gender',
    'Currency']
campaign_df = DataFrame(
campaign_data.tolist(), columns=campaign_cols, index=df.index)
campaign_df.head(3)
```

Así queda mejor. Una cosa importante a recordar: incluso si las fechas se imprimen como cadenas, son sólo la representación de los objetos de fecha reales que se almacenan en `DataFrame`.

Otra cosa muy importante: al unir dos instancias de `DataFrame`, es imprescindible que tengan el mismo índice, de lo contrario `pandas` no podrá saber qué filas van con cada una. Por lo tanto, cuando creamos `campaign_df`, establecemos su índice al de `df`. Esto nos permite unirlas:

```python
df = df.join(campaign_df)
```

Y después de `join()`, echamos un vistazo, esperando ver datos coincidentes:

```python
df[['cmp_name'] + campaign_cols].head(3)
```

Como puede ver, `join()` tuvo éxito; el nombre de la campaña y las columnas separadas exponen los mismos datos. Estamos accediendo al `DataFrame` usando la sintaxis de corchetes, y pasamos una lista de nombres de columnas. Esto producirá un nuevo `DataFrame`, con esas columnas (en el mismo orden), sobre el que luego llamamos al método `head()`.

### Descomponer los datos del usuario
Ahora hacemos lo mismo para cada pieza de datos JSON del usuario. Llamamos a `apply()` en la serie de usuarios, ejecutando la función `unpack_user_json()`, que toma un objeto JSON de usuario y lo transforma en una lista de sus campos. Creamos un nuevo `DataFrame`, `user_df`, con estos datos y volvemos a unir `user_df` con `df` (como hicimos con `campaign_df`) y el resultado es el siguiente:

```python
def unpack_user_json(user):
    '''
    Muy optimista también, espera que los objetos de usuario tengan 
    todos los atributos
    '''
    user = json.loads(user.strip())
    return [
        user['username'],
        user['email'],
        user['name'],
        user['gender'],
        user['age'],
        user['address'],
    ]

user_data = df['user'].apply(unpack_user_json)

user_cols = [
    'Username', 'Email', 'Name', 'Gender', 'Age', 'Address']

user_df = DataFrame(
    user_data.tolist(), columns=user_cols, index=df.index)

df = df.join(user_df)

df[['user'] + user_cols].head(2)
```

El resultado nos muestra que todo salió bien. Sin embargo, todavía no hemos terminado. Si llamas a `df.columns` en una celda, verás que aún tenemos nombres feos para nuestras columnas. Cambiemos eso:

```python
new_column_names = {
    'cmp_bgt': 'Budget',
    'cmp_spent': 'Spent',
    'cmp_clicks': 'Clicks',
    'cmp_impr': 'Impressions',
}

df.rename(columns=new_column_names, inplace=True)
```

El método `rename()` se puede utilizar para cambiar las etiquetas de las columnas (o filas). Le hemos dado un diccionario que asigna los nombres antiguos de las columnas a nuestros nombres preferidos. Cualquier columna que no se mencione en el diccionario permanecerá sin cambios. Ahora, con la excepción de 'cmp_name' y 'user', sólo tenemos nombres bonitos.

Nuestro siguiente paso será añadir algunas columnas adicionales. Para cada campaña, tenemos el número de clics e impresiones, y tenemos las cantidades gastadas. Esto nos permite introducir tres ratios de medición: CTR, CPC y CPI. Son las siglas de Click Through Rate (porcentaje de clics), Cost Per Click (coste por clic) y Cost Per Impression (coste por impresión), respectivamente.

Los dos últimos son sencillos, pero el CTR no lo es. Basta con decir que es la relación entre clics e impresiones. Te da una medida de cuántos clics se realizaron en un anuncio de una campaña por impresión: cuanto mayor sea este número, más éxito tendrá el anuncio a la hora de atraer a los usuarios para que hagan clic en él. Escribamos una función que calcule los tres ratios y los añada al `DataFrame`:

```python
def calculate_extra_columns(df):
    
    # Click Through Rate
    df['CTR'] = df['Clicks'] / df['Impressions']

    # Cost Per Click
    df['CPC'] = df['Spent'] / df['Clicks']

    # Cost Per Impression
    df['CPI'] = df['Spent'] / df['Impressions']

calculate_extra_columns(df)
```

Observa que estamos añadiendo esas tres columnas con una línea de código cada una, pero el DataFrame aplica la operación automáticamente (la división, en este caso) a cada par de celdas de las columnas correspondientes. Así, aunque parezca que sólo estamos haciendo tres divisiones, en realidad hay tres veces el total de valores de la base de datos, porque se realizan para cada fila. `pandas` hace mucho trabajo por nosotros, y además hace un muy buen trabajo ocultando la complejidad del mismo.

La función `calculate_extra_columns()` toma un `DataFrame (df)`, y trabaja directamente sobre él. Este modo de operación se denomina in-place. Es similar a como el método `list.sort()` ordena una lista. También se podría decir que esta función no es pura, lo que significa que tiene efectos secundarios, ya que modifica el objeto mutable que se le pasa como argumento.

Podemos echar un vistazo a los resultados filtrando en las columnas relevantes y llamando a `head()`:

```python
df[['Spent', 'Clicks', 'Impressions',
    'CTR', 'CPC', 'CPI']].head(3)
```

Ahora queremos verificar manualmente la exactitud de los resultados de la primera fila, cuyos resultados son:

```python
clicks = df['Clicks'][0]
impressions = df['Impressions'][0]
spent = df['Spent'][0]
CTR = df['CTR'][0]
CPC = df['CPC'][0]
CPI = df['CPI'][0]
print('CTR:', CTR, clicks / impressions)
print('CPC:', CPC, spent / clicks)
print('CPI:', CPI, spent / impressions)
```

Los valores coinciden, lo que confirma que nuestros cálculos son correctos. Por supuesto, normalmente no se necesita hacer esto, pero es una manera de mostrar lo que internamente realiza `pandas`. Puede acceder a una `Serie` (una columna) pasando su nombre al `DataFrame` entre corchetes (esto es similar a buscar una clave en un diccionario). A continuación, puede acceder a cada fila de la columna por su posición, exactamente como lo haría con una lista normal o tupla.

Ya casi hemos terminado con nuestro `DataFrame`. Lo único que nos falta ahora es una columna que nos diga la duración de la campaña y una columna que nos diga qué día de la semana corresponde a la fecha de inicio de cada campaña. Es importante tener la duración, ya que nos permite relacionar datos como la cantidad gastada o el número de impresiones con la duración de la campaña (podemos esperar que las campañas de mayor duración cuesten más y tengan más impresiones). El día de la semana también puede ser útil; por ejemplo, algunas campañas pueden estar vinculadas a acontecimientos que tienen lugar en días concretos de la semana (como eventos deportivos que se celebran los sábados o los domingos).

```python
def get_day_of_the_week(day):
    return day.strftime("%A")

def get_duration(row):
    return (row['End'] - row['Start']).days

df['Day of Week'] = df['Start'].apply(get_day_of_the_week)

df['Duration'] = df.apply(get_duration, axis=1)
```

`get_day_of_the_week()` toma un objeto `date` y le da formato de string, que sólo contiene el nombre del día de la semana correspondiente. `get_duration()`; en primer lugar, toma una fila entera, no sólo un valor. En su cuerpo, se realiza una resta entre las fechas de fin e inicio de una campaña. Al restar objetos `date`, el resultado es un objeto `timedelta`, que representa una determinada cantidad de tiempo. Tomamos el valor de su propiedad `.days` para obtener la duración en días.

Ahora, podemos introducir la parte divertida, la aplicación de esas dos funciones. Primero, aplicamos `get_day_of_the_week()` a la columna Start (como objeto `Series`); esto es similar a lo que hicimos con 'user' y 'cmp_name'. A continuación, aplicamos `get_duration()` a todo el `DataFrame` y, para indicar a `pandas` que realice esa operación sobre las filas, le pasamos `axis=1`.

Podemos comprobar los resultados muy fácilmente, como se muestra aquí:

```python
df[['Start', 'End', 'Duration', 'Day of Week']].head(3)
```

Ahora sabemos que entre el 24 de abril de 2024 y el 20 de noviembre de 2025 hay 575 días y que el 03 de julio de 2025 es jueves.

En este caso, preparamos el `selector` para filtrar todas las filas cuyo `spend` sea inferior o igual al 75% del `budget`. En otras palabras, incluiremos sólo aquellas campañas para las que hayamos gastado al menos tres cuartas partes del presupuesto. Fíjese que en el selector le estamos mostrando una forma alternativa de pedir una columna del `DataFrame`, utilizando el acceso directo a la propiedad (`object.property_name`), en lugar del acceso tipo diccionario (`object['property_name']`). Si `property_name` es un nombre Python válido, puedes utilizar ambas formas indistintamente.

El `selector` se aplica de la misma forma que accedemos a un diccionario con una clave. Cuando aplicamos `selector` a `df`, obtenemos de vuelta un nuevo `DataFrame` y seleccionamos sólo las columnas relevantes de éste y volvemos a llamar a `hist()`.

Observe que las formas de los gráficos no han cambiado mucho, salvo la del gráfico `Spent`, que es bastante diferente. Esto significa que sólo incluimos las líneas en las que el importe gastado se aproxima al presupuesto. Las cifras del presupuesto proceden de una distribución uniforme. Por lo tanto, es bastante obvio que el gráfico `Spent` adopta ahora ese tipo de forma. Pidamos ahora algo diferente. ¿Qué tal la medida de `Spend`, `Clicks` e `Impressions` agrupadas por día de la semana?

```python
# Define el orden correcto de los días de la semana
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Convierte la columna 'Day of Week' en una categoría con el orden correcto
df['Day of Week'] = pd.Categorical(df['Day of Week'], categories=day_order, ordered=True)

# Agrupa y suma solo las columnas numéricas
df_weekday = df.groupby('Day of Week', observed=True)[['Impressions', 'Spent', 'Clicks']].sum()

# Graficar los resultados
df_weekday.plot(figsize=(16, 6), subplots=True);
```

La primera línea crea un nuevo `DataFrame`, `df_weekday`, solicitando una agrupación por `Day of Week` en `df` de las columnas `Impressions`, `Spent` y `Clicks`. La función que se utiliza para agregar los datos es `sum()`.

La segunda línea, llamamos a `plot()`, que es un poco diferente a `hist()`. La opción `subplots=True` hace que plot dibuje tres gráficos separados.

Curiosamente, podemos ver que la mayor parte de la acción tiene lugar los viernes y los domingos. Si se tratara de datos significativos, podría ser una información importante para nuestros clientes, y por eso les mostramos este ejemplo.

Terminamos este apartado con un par de cosas más. Primero, una simple agregación. Queremos agregar en `Target Gender` y `Target Age`, y mostrar `Impressions` y `Spend`. Para ambos, queremos ver la `mean` y la desviación estándar (`std`).

Es muy fácil de hacer. Preparamos un diccionario que utilizaremos como configuración. A continuación, realizamos una agrupación en las columnas `Target Gender` y `Target Age`, y pasamos nuestro diccionario de configuración al método `agg()`. El resultado es el siguiente:

```python
agg_config = {
    'Impressions': ['mean', 'std'],
    'Spent': ['mean', 'std'],
}

df.groupby(['Target Gender', 'Target Age']).agg(agg_config)
```

Vamos a hacer una cosa más. Queremos mostrarte algo llamado tabla dinámica (pivot table). Una tabla dinámica es una forma de agrupar datos, calcular un valor agregado para cada grupo y mostrar el resultado en forma de tabla. La tabla dinámica es una herramienta esencial para el análisis de datos, así que veamos un ejemplo sencillo.

Vamos a crear una tabla dinámica que nos muestra la correlación entre `Target Age` e `Impressions`, `Clicks` y `Spend`. Estos tres últimos se subdividirán en función de `Target Gender`. La función de agregación (`aggfunc`) utilizada para calcular los resultados es la función `numpy.sum()` (`numpy.mean` sería la función por defecto, si no hubiéramos especificado nada). El resultado es un nuevo `DataFrame` que contiene la tabla pivotante:

```python
df.pivot_table(
    values=['Impressions', 'Clicks', 'Spent'],
    index=['Target Age'],
    columns=['Target Gender'],
    aggfunc=np.sum
)
```

El resultado es bastante claro y proporciona información muy útil cuando los datos son significativos.
